```python
from ECG.ecg import read_ecg
from ECG.ecg_sig import butter_highpass_filter,butter_lowpass_filter
import os
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')


from scipy.signal import correlate


events = 30000
eventsr = 4*events
# file  =  os.path.join(os.path.dirname(os.path.realpath(__file__)),"a01.dat")
# filer  =  os.path.join(os.path.dirname(os.path.realpath(__file__)),"a01r.dat")
y = read_ecg(file,0,events)
yr = read_ecg(filer,0,eventsr)

respN = yr[2::4]

fps = 100

Filteredr = butter_highpass_filter(respN,0.1,fps)
Filteredr = butter_lowpass_filter(Filteredr,0.2,fps)

Filterede = butter_highpass_filter(y,0.1,fps)
Filterede = butter_lowpass_filter(Filterede,0.2,fps)
```

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
df = pd.DataFrame({"Fe":Filterede,
             "Fr":Filteredr})
df.plot(subplots=True)

num_pipeline = Pipeline([
            ('std_scaler', StandardScaler()),
        ])

scale = num_pipeline.fit_transform(df)
df_scale = pd.DataFrame(scale,columns=["Fe","Fr"])

for i in range(1,400):
    df_scale[f"s{i}"] = df_scale["Fe"].shift(i)
```

```python
corr_matrix = df_scale.corr()
```

```python
fe_corr = corr_matrix["Fr"][2:]
# print(np.argmax(fe_corr),np.max(fe_corr))
```

```python
print(fe_corr.idxmax(),fe_corr.max())
```

```python
df_re = pd.DataFrame({"X":df_scale["s138"],"y":df_scale["Fr"]})
df_re = df_re.dropna()
df_re.head()

```

```python
from sklearn.svm import SVR


# X = df_re["X"].reshape(-1, 1)
# y = df_re["y"].reshape(-1, 1)
X = df_re["X"].values.reshape(-1, 1)
y = df_re["y"].values.reshape(-1, 1)
clf = SVR()
clf.fit(X, y) 


# yp = regr.predict(X)
# print(regr.predict([[0, 0]])) 
```

```python
clf.score(X,y)
yp = clf.predict(X)
```
