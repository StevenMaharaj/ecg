```python
from ECG.ecg import read_ecg, make_df_summary
import os
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [16, 8]
from scipy.signal import find_peaks
from scipy import fft
import numpy as np
import pandas as pd
```

```python
# find length of data

with open("a01.dat", "rb") as binary_file:
    # Read the whole file at once
    n = len(binary_file.read())
    
```

```python
# find correlations

events = n
# file  =  os.path.join(os.path.dirname(os.path.realpath(__file__)),"a01.dat")
file = "a01.dat"
df = make_df_summary(file,0,events=n)
corr = df[1:].corr()
sns.heatmap(corr, annot=True, fmt=".2f",cmap="BuPu")

sns.pairplot(df[1:],kind="reg",plot_kws={'line_kws':{'color':'blue'}, 'scatter_kws': {'alpha': 0.1}})
```

```python
plt.style.use('ggplot')

events = 800
# file  =  os.path.join(os.path.dirname(os.path.realpath(__file__)),"a01r.dat")
file = "a01r.dat"
yr = read_ecg(file,0,events)


fig,ax = plt.subplots(nrows=4,ncols = 1)
for i in range(3):
    ax[i].plot(yr[i::4])

plt.tight_layout()
plt.show()
```

```python
events = 1000
eventsr = 4*events
# file  =  os.path.join(os.path.dirname(os.path.realpath(__file__)),"a01.dat")
# filer  =  os.path.join(os.path.dirname(os.path.realpath(__file__)),"a01r.dat")

file = "a01.dat"
filer = "a01r.dat"
y = read_ecg(file,0,events)
Respc = read_ecg(filer,0,eventsr)

fig,ax = plt.subplots(2,1)

ax[0].plot(y)
ax[0].set_title(f"ECG for {events} samples or {events/100} secs ")
# ax[0].set_xlabel("200 A/D units per millivolt")
ax[1].plot(Respc[::4])
ax[1].set_title(f"Resp C for {events} samples or {events/100} secs ")
plt.figure()
plt.plot(np.abs(fft(Respc)))
plt.title("FFT of Respc")
plt.tight_layout()
plt.show()
```
